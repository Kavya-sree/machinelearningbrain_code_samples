```python
# Importing Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading sample dataset from seaborn
df = sns.load_dataset('penguins')
df.head()
```

# Using Matplotlib

## [matplotlib.pyplot.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

You can use .plot() of matplotlib. It plots y versus x as lines and/or markers. By default, the plot() function draws a line from point to point. But if you specify a marker, we get a scatterplot. Use any of the string notation parameter for marker. For example to get circle marker use notation "o". You can find other list of markers in this [link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

```python
plt.plot(df["bill_length_mm"], df["bill_depth_mm"], "o")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

```python
plt.plot(df["bill_length_mm"], df["bill_depth_mm"], "p", color='red')
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

## [matplotlib.pyplot.scatter()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)


```python
plt.scatter(df["bill_length_mm"], df["bill_depth_mm"])
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

### Customization

* Shape- The default marker is a solid round shape. To change the shape or style of the marker, specify the required marker in the `marker` parameter.
* Size- You can also change the size of the marker. Just specify marker size parameter `s`.
* Color- We can also change the colour of the marker. The default color is blue. You can choose any color.
* Edge color- We can change the edge colour of the marker using the argument "edgecolour". We can also set the line width of the edge using "linewidth".
* Transparency- The `alpha` argument of the plot controls the transparency of the markers. The alpha blending value is between 0 (transparent) and 1 (opaque). This could be useful when there are a lot of markers together, so you will be able to see the high-density regions.

The following code contains each of these customization:

```python
# setting marker shape as 'v' for traingle_down shape, color as 'green' and size as 60
plt.scatter(df["bill_length_mm"], df["bill_depth_mm"], marker = 'v', c= 'green', s=60, alpha=0.5)
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

```python
# setting marker shape as 'p' for pentagon shape, color as 'red' and size as 60, edgecolor as 'yellow', and linewidth as 1.5
plt.scatter(df["bill_length_mm"], df["bill_depth_mm"], marker = 'p', c= 'red', s=60, edgecolor='yellow', linewidth=1.5, alpha=0.6)
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

### Color by category

```python
# Scatterplot by category "species"
colors = {'Adelie':'red', 'Chinstrap':'green', 'Gentoo':'blue'}

species_group = df.groupby("species")
for key, group in species_group:
  plt.scatter(group.bill_length_mm, group.bill_depth_mm, marker = 'p', label = key, color=colors[key])

plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.legend()
plt.show()
```

# Using Seaborn

```python
# Custom Color Palletes
my_colors = ['#ea7d23', '#f9e94e', '#66b2b2']

plt.figure(figsize=(12,6))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=df , hue= "species", style = "species", palette = my_colors, markers=['o', 's', 'X'])
plt.suptitle('Scatterplot of bill length in relation to bill depth', fontweight = 'heavy', y = 1.03, fontsize = '16')
plt.show()
```
